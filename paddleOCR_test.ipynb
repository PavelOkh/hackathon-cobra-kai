# используем PaddleOCR

```python
ocr = PaddleOCR(use_angle_cls=True, lang='en')
```

```python
img_path = '/content/X51008099090.jpg' #указать путь к файлу
img = cv2.imread(img_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
result = ocr.ocr(img, cls=True)
for idx in range(len(result)):
    res = result[idx]
    for line in res:
        print(line)
```

# Сохраняем картинку с обведенными распознанными ячейками 

```python
output_path = '/content/output_img/'
os.mkdir(output_path)
```

```python
from PIL import Image
res = result[0]
image = Image.open(img_path).convert('RGB')
boxes = [line[0] for line in res]
txts = [line[1][0] for line in res]
scores = [line[1][1] for line in res]
im_show = draw_ocr(image, boxes)
im_show = Image.fromarray(im_show)
im_show.save('/content/output_img/result.jpg')
im_show
```

# Сохраняем в файл txt координаты и распознанные слова

```python
with open(output_path+"result.txt", "w") as external_file:
    for idx in range(len(result)):
        res = result[idx]
        for line in res:
            line_flatten = list(flatten(line))
            print(*line_flatten[:-1], sep =', ', file=external_file)
            print(*line_flatten[:-1], sep =', ')
    external_file.close()
```
